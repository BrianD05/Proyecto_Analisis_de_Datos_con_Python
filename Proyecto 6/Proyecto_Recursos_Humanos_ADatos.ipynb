# Distribución de la edad de los empleados

•	¿Cómo se distribuye la edad de los empleados en la empresa?

•	¿Hay alguna correlación entre la edad y el rendimiento de los empleados?


```python
# Convierto la columna ''Fecha_Nacimiento' a tipo fecha y calcular la edad de los empleados
df['Fecha_Nacimiento'] = pd.to_datetime(df['Fecha_Nacimiento'])
df['Edad'] = (pd.to_datetime('now') - df['Fecha_Nacimiento']).astype('<m8[Y]')
```

```python
#Elimino filas de edades superiores a la actual ya que hay incosistencia
# Filtro las fechas que sean mayores al 11 de febrero de 2024 y obtiene los índices de esas filas
indices_fechas_superiores = df[df['Fecha_Nacimiento'] > '2024-02-11'].index

# Elimino las filas con fechas superiores al 11 de febrero de 2024
df = df.drop(indices_fechas_superiores)

df
```

```python
distribucion_edad_empleados = df['Edad'].value_counts()
print("Distribucion de Edad de los Empleados: ")
print(distribucion_edad_empleados)
```

```python
import matplotlib.pyplot as plt
# Necesito ordenar los datos antes de trazarlos
distribucion_edad_empleados_ordenados = distribucion_edad_empleados.sort_values(ascending=False)

# Trazar los datos ordenados
distribucion_edad_empleados_ordenados.plot(kind='bar')

plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribuciones de Edades')
plt.show()
```

```python
# Calcular la correlación entre la edad y el rendimiento de los empleados
correlacion = df['Edad'].corr(df['ID_Rendimiento'])
print("Correlación entre la edad y el rendimiento de los empleados:", correlacion)
```

# Distribución de género en la empresa:

•	¿Cuál es la proporción de hombres y mujeres en la empresa?


```python
#Proporcion de los generos
proporcion_genero_empresa = df['Género'].value_counts(normalize = True)
print("Proporción de género en la empresa")
print(proporcion_genero_empresa)
```

```python
import matplotlib.pyplot as plt
# Visualiza los resultados (gráfico circular)
plt.figure(figsize = (12,8))
plt.pie(proporcion_genero_empresa,labels = proporcion_genero_empresa.index,autopct = '%1.1f%%',startangle=140)
plt.title('Proporción de los Empleados')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

El 54,7 % son del Género Femenino mientras que el 45,3% son del Género Masculino

•	¿Cómo se distribuyen los diferentes géneros en cada departamento?

```python
distribucion_genero_departamento = df.groupby(['Departamento'])['Género'].value_counts()

# Pivota los datos para tener los géneros como columnas
distribucion_genero_departamento_tabla = distribucion_genero_departamento.unstack()

# Imprimo la tabla
print("\nDistribución de género en cada departamento:")
print(distribucion_genero_departamento_tabla)
```

```python
# Creo un gráfico de barras
plt.figure(figsize=(9, 8))  # Establecer el tamaño del gráfico

# Uso el método unstack() para pivotar los datos y tener los géneros como columnas
distribucion_genero_departamento.unstack().plot(kind='bar', stacked=True)

# Configuración del gráfico
plt.title('Distribución de género en cada departamento')
plt.xlabel('Departamento')
plt.ylabel('Número de empleados')
plt.xticks(rotation=45, ha='right')  # Rota las etiquetas del eje x para mejor visualización

# Mostrar el gráfico
plt.legend(title='Género')
plt.tight_layout()  # Ajusta el diseño del gráfico
plt.show()
```

# Salario

•	¿Existe alguna relación entre el salario y el desempeño de los empleados?

•	¿Cómo varía el salario promedio en función del desempeño?


```python
# Agrupa los datos por el nivel de desempeño y calcula el salario promedio para cada grupo
salario_promedio_por_desempeno = df.groupby('ID_Rendimiento')['Salario'].mean().sort_values(ascending = False)

# Imprimo el salario promedio por nivel de desempeño
print("Salario promedio por nivel de desempeño:")
print(salario_promedio_por_desempeno)
```

```python
# Grafica el salario promedio por nivel de desempeño
salario_promedio_por_desempeno.plot(kind='bar')
plt.xlabel('Nivel de Desempeño')
plt.ylabel('Salario Promedio')
plt.title('Salario Promedio por Nivel de Desempeño')
plt.show()
```

# Evaluación del compromiso y satisfacción laboral
•	¿Cómo se distribuyen las puntuaciones de compromiso y satisfacción laboral entre los empleados?

•	¿Hay alguna correlación entre el compromiso y la satisfacción laboral?


```python
df.columns
```

```python
# Calculo la distribución de las puntuaciones de compromiso y satisfacción laboral
distribucion_empleados = df.groupby(['Puntuación_Rendimiento', 'Satisfacción_Empleado']).size().unstack(fill_value=0)

# Agrega un total por fila
distribucion_empleados['Total'] = distribucion_empleados.sum(axis=1)

# Muestra la tabla
print("Distribución de Puntuaciones de Compromiso y Satisfacción Laboral:")
print(distribucion_empleados)
```

```python
# Elimino la columna 'Total' para evitar que se incluya en la gráfica
distribucion_empleados_sin_total = distribucion_empleados.drop('Total', axis=1)

# Genero la gráfica de barras
distribucion_empleados_sin_total.plot(kind='bar', stacked=True, figsize=(10, 6))

# Ajusto las etiquetas y el título
plt.title('Distribución de Puntuaciones de Compromiso y Satisfacción Laboral')
plt.xlabel('Puntuación de Satisfacción')
plt.ylabel('Número de Empleados')
plt.xticks(rotation=0)  # Rotación de etiquetas en el eje x

# Muestro la leyenda fuera del gráfico
plt.legend(title='Puntuación de Rendimiento', bbox_to_anchor=(1.05, 1), loc='upper left')

# Muestro la gráfica
plt.tight_layout()
plt.show()
```

```python
#¿Hay alguna correlación entre el compromiso y la satisfacción laboral?
correlacion = df['Encuesta_Compromiso'].corr(df['Satisfacción_Empleado'])
print("Correlación entre el Compromiso y la satisfacion Laboral:", correlacion)
```

# Origen de reclutamiento y desempeño
•	¿Cuáles son las fuentes de reclutamiento más comunes?

•	¿Cuántos empleados reclutados a través de Indeed y LinkedIn comenzaron y terminaron su empleo en la empresa?


```python
fuente_reclutamiento = df['Fuente_Reclutamiento'].value_counts()

print("Fuentes de reclutamiento mas comunes")
print(fuente_reclutamiento)
```

```python
# Filtro empleados reclutados a través de Indeed y LinkedIn
empleados_indeed = df[(df['Fuente_Reclutamiento'] == 'Indeed') & (df['Fecha_Contratación'].notnull()) & (df['Fecha_Terminación'].notnull())]
empleados_linkedin = df[(df['Fuente_Reclutamiento'] == 'LinkedIn') & (df['Fecha_Contratación'].notnull()) & (df['Fecha_Terminación'].notnull())]

# Cuántos empleados reclutados a través de Indeed y LinkedIn comenzaron y terminaron su empleo
cantidad_indeed = len(empleados_indeed)
cantidad_linkedin = len(empleados_linkedin)

print(f"Empleados reclutados a través de Indeed que comenzaron y terminaron su empleo: {cantidad_indeed}")
print(f"Empleados reclutados a través de LinkedIn que comenzaron y terminaron su empleo: {cantidad_linkedin}")

```

# Frecuencia de ausencias


•	¿Cuál es la frecuencia promedio de ausencias de los empleados?


```python
frecuencia_ausencia = df['Ausencias'].mean()

print("El Promedio de Ausencia es: ",frecuencia_ausencia)
```

```python
#Dato Adicional sobre la ausencias de los empleados
personas_ausentes = df.groupby('Nombre_Empleado')['Ausencias'].value_counts().reset_index(name='Cantidad')

print(personas_ausentes)
```

El DataFrame `personas_ausentes` muestra la cantidad de ausencias para cada empleado en el conjunto de datos.

- En la columna `Nombre_Empleado`, se encuentran los nombres de los empleados.
- La columna `Ausencias` indica el número de días de ausencia para cada empleado.
- La columna `Cantidad` muestra cuántas veces ocurrió ese número específico de ausencias para cada empleado.

Por ejemplo:
- Adinolfi, Wilson K tuvo 1 ausencia de 1 día.
- Ait Sidi, Karthikeyan tuvo 1 ausencia de 17 días.
- Akinkuolie, Sarah tuvo 1 ausencia de 3 días.
- Alagbe, Trina tuvo 1 ausencia de 15 días.
- Anderson, Carol tuvo 1 ausencia de 2 días.
