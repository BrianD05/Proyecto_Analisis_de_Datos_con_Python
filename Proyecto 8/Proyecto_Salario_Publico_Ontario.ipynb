```python
import os
import pandas as pd

# Obtén el directorio actual
directorio_actual = os.getcwd()

# Lista de nombres de archivos CSV
nombres_archivos = ["ontario-public-sector-salary-2016.csv", "ontario-public-sector-salary-2017.csv", "ontario-public-sector-salary-2018.csv", "ontario-public-sector-salary-2019.csv"]

# Lista para almacenar los DataFrames de cada archivo CSV
dataframes = []

# Contador para iterar sobre los nombres de archivo
indice = 0

# Mientras el índice sea menor que la longitud de la lista de nombres de archivos
while indice < len(nombres_archivos):
    # Construye la ruta completa del archivo utilizando os.path.join
    ruta_completa = os.path.join(directorio_actual, nombres_archivos[indice])
    
    # Verifica si el archivo existe antes de intentar abrirlo
    if os.path.exists(ruta_completa):
        # Abre el archivo CSV utilizando pandas y lo agrega a la lista de DataFrames
        df = pd.read_csv(ruta_completa)
        dataframes.append(df)
    else:
        print(f"El archivo {nombres_archivos[indice]} no existe en el directorio actual.")
    
    # Incrementa el índice para pasar al siguiente archivo
    indice += 1

# Concatena todos los DataFrames en uno solo
df_completo = pd.concat(dataframes, ignore_index=True)
```

Salary Paid (Salario Pagado):

count: Es el número de registros no nulos en la columna de salarios pagados.

mean: Es el promedio de los salarios pagados.

std: Es la desviación estándar de los salarios pagados, que indica qué tan dispersos están los salarios alrededor de la media.

min: Es el valor mínimo de los salarios pagados.

25%: Es el percentil 25, es decir, el valor por debajo del cual cae el 25% de los salarios pagados.

50%: Es el percentil 50, que es equivalente a la mediana y representa el valor por debajo del cual cae el 50% de los salarios pagados.

75%: Es el percentil 75, es decir, el valor por debajo del cual cae el 75% de los salarios pagados.

max: Es el valor máximo de los salarios pagados.

Taxable Benefits (Beneficios Imponibles):

La interpretación de los estadísticos es similar a la de "Salary Paid". Estos estadísticos describen la distribución de los beneficios imponibles.

Calendar Year (Año Calendario):

count: Es el número de registros no nulos en la columna del año calendario.

mean: Es el promedio de los años calendario.

std: Es la desviación estándar de los años calendario.

min: Es el valor mínimo de los años calendario.

25%: Es el percentil 25 de los años calendario.

50%: Es el percentil 50 de los años calendario.

75%: Es el percentil 75 de los años calendario.

max: Es el valor máximo de los años calendario.

¿Cuál es la distribución de los sectores representados en el conjunto de datos?

```python
distribucion_sector = df_completo['Sector'].value_counts()
print('La Distribucion es: \n',distribucion_sector)
```

```python
#Visualizo los datos en una grafica de barras
import matplotlib.pyplot as plt
# Necesito ordenar los datos antes de trazarlos
distribucion_sector_ordenados = distribucion_sector.sort_values(ascending=False)
# Trazar los datos ordenados
distribucion_sector_ordenados.plot(kind='bar',figsize = (12,6))
plt.xlabel('Sector')
plt.ylabel('Frecuencia')
plt.title('Distribuciones de Sectores')
plt.show()
```

¿Cuál es el salario más alto pagado a un empleado en el conjunto de datos?

```python
salario_maximo = df_completo['Salary Paid'].max()
print("El salario Maximo es: ",salario_maximo)
```

```python
#Dato Adicional Buscaremos al empleado que tiene el salario mas alto
# Encuentro el índice del registro con el salario máximo
indice_salario_maximo = df_completo['Salary Paid'].idxmax()

# Utilizo el índice para obtener el nombre de la persona con el salario máximo
persona_salario_maximo = df_completo.loc[indice_salario_maximo, 'First Name']

print("La persona con el salario máximo es:", persona_salario_maximo,"y su Sueldo es",salario_maximo,"Dolares")
```

```python
#Los 10 mejores sueldos segun su nombre apellido salario y el año del salario
# Ordeno el DataFrame por el salario pagado de manera descendente y selecciona los primeros 10 registros
mejores_sueldos = df_completo.sort_values(by='Salary Paid', ascending=False).head(10)

# Imprime los 10 mejores sueldos con sus nombre apellido salario y el año del salario correspondientes
print(mejores_sueldos[['First Name', 'Last Name','Salary Paid','Calendar Year']])
```

```python
#Visualizo con un grafico circular en porcentajes obtenidos de los 10 mejores salarios
#Dato Adicional: grafico circular para tener una visualizacion de los datos en porcentajes de salarios
# Visualiza los resultados (gráfico circular)
plt.figure(figsize=(10, 5))
plt.pie(mejores_sueldos['Salary Paid'], labels=mejores_sueldos['Last Name'] + ', ' + mejores_sueldos['First Name'] + ', ' + mejores_sueldos['Calendar Year'].astype(str), autopct='%1.1f%%', startangle=140)
plt.suptitle('Los 10 Mejores Sueldos  de los Empleados Anualmente', y=1, fontsize=20)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
