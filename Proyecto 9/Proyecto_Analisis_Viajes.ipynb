Objetivos del Proyecto:

Desvelar la Duración Promedio de los Viajes: Investigar la duración media de los viajes registrados para comprender mejor las preferencias de duración de los viajeros.

Explorar la Relación entre Duración del Viaje y Costo del Alojamiento: Investigar posibles relaciones entre la duración del viaje y el gasto en alojamiento para revelar patrones de gasto.

Analizar la Distribución de Viajeros por Género: Explorar la distribución de género entre los viajeros, proporcionando información sobre la diversidad de género en los viajes.

Identificar el Destino con el Alojamiento más Costoso: Revelar cuál es el destino con el alojamiento más caro, destacando las opciones de hospedaje exclusivas.

Identificar el Destino con el Transporte más Costoso: Determinar el destino con el transporte más oneroso, ofreciendo información sobre las opciones de transporte lujosas.

Explorar la Relación entre Edad de los Viajeros y Costo Total del Viaje: Investigar si la edad de los viajeros influye en el costo total del viaje, revelando tendencias generacionales en los gastos de viaje.

Identificar el Destino más Costoso en Términos de Costo Total del Viaje: Determinar el destino con el costo total de viaje más elevado, considerando tanto alojamiento como transporte.

Identificar el Destino más Económico en Términos de Costo Total del Viaje: Descubrir el destino más asequible en términos de costo total de viaje, proporcionando opciones para viajeros con presupuestos ajustados.

Calcular el Medio de Transporte más Costoso en Promedio: Determinar cuál es el medio de transporte con el costo más elevado en promedio, ofreciendo información sobre las preferencias y costos asociados con cada modo de transporte.

# Importacion del Archivo

```python
import os
import pandas as pd
# Obtiene el directorio actual
directorio_actual = os.getcwd()

# Nombre del archivo CSV que se desea abrir
nombre_archivo = "Travel details dataset_trips.csv"

# Se Construye la ruta completa del archivo utilizando os.path.join
ruta_completa = os.path.join(directorio_actual, nombre_archivo)

# Verifico si el archivo existe antes de intentar abrirlo
if os.path.exists(ruta_completa):
    # Abre el archivo CSV utilizando pandas
    df = pd.read_csv(ruta_completa)

else:
    print(f"El archivo {nombre_archivo} no existe en el directorio actual.")
```

¿Hay alguna relación entre la duración del viaje y el costo del alojamiento?

```python
# Calcular la correlación entre las variables
correlacion = df['Duration (days)'].corr(df['Accommodation cost'])

# Creo un gráfico de dispersión para visualizar la relación
plt.figure(figsize=(10, 6))
plt.scatter(df['Duration (days)'], df['Accommodation cost'], color='blue', alpha=0.5)
plt.title('Relación entre Duración del Viaje y Costo del Alojamiento')
plt.xlabel('Duración del Viaje (días)')
plt.ylabel('Costo del Alojamiento')
plt.grid(True)
plt.show()
print(f"La correlación entre la duración del viaje y el costo del alojamiento es: {correlacion}")
```

La correlación entre la duración del viaje y el costo del alojamiento es aproximadamente -0.127. Esta correlación negativa sugiere que, en general, no hay una relación lineal fuerte entre la duración del viaje y el costo del alojamiento

¿Cuál es el destino con el transporte más costoso?

¿Hay alguna relación entre la edad de los viajeros y el costo total del viaje?

```python
# Calcula el costo total del viaje sumando el costo del alojamiento y el costo del transporte
df['Total trip cost'] = df['Accommodation cost'] + df['Transportation cost']

# Calcula la correlación entre la edad de los viajeros y el costo total del viaje
correlacion = df['Traveler age'].corr(df['Total trip cost'])

print(f"La correlación entre la edad de los viajeros y el costo total del viaje es: {correlacion}")
```

La correlación entre la edad de los viajeros y el costo total del viaje es de 0.0379, lo que sugiere una relación muy débil o prácticamente inexistente entre ambas variables. No hay una asociación clara entre la edad de los viajeros y cuánto gastan en total en su viaje, al menos en el conjunto de datos analizado. 


¿Cuál es el destino más costoso en términos de costo total del viaje (alojamiento + transporte)?

```python
# Encuentro el destino más costoso
destino_mas_costoso = df.loc[df['Total trip cost'].idxmax(), 'Destination']
costo_maximo = df['Total trip cost'].max()

print(f"El destino más costoso es '{destino_mas_costoso}' con un costo de {costo_maximo} USD.")
```

¿Cuál es el destino más económico en términos de costo total del viaje?

```python
# Encuentro el destino más economico
destino_mas_economico = df.loc[df['Total trip cost'].idxmin(), 'Destination']
costo_minimo = df['Total trip cost'].min()

print(f"El destino más costoso es '{destino_mas_economico}' con un costo de {costo_minimo} USD.")
```

¿Cuál es el medio de transporte más costoso en promedio?

```python
# Calcula el costo promedio del transporte para cada tipo
costo_promedio_transporte = df.groupby('Transportation type')['Transportation cost'].mean()

# Ordena los resultados de mayor a menor
costo_promedio_transporte_ordenado = costo_promedio_transporte.sort_values(ascending=False)

# Muestra los resultados ordenados
print(costo_promedio_transporte_ordenado)
```

```python
import numpy as np
# Genera una paleta de colores aleatorios
num_transporte = len(costo_promedio_transporte_ordenado)
colores = np.random.rand(num_transporte, 3)  # Genera colores RGB aleatorios

# Traza los datos con colores aleatorios
costo_promedio_transporte_ordenado.plot(kind='bar', figsize=(12, 6), color=colores)

plt.xlabel('Tipo de Transporte')
plt.ylabel('Costo Promedio')
plt.title('Costo Promedio por Tipo de Transporte')
plt.xticks(rotation=45, ha='right')  # Rota las etiquetas del eje x para mejor legibilidad
plt.tight_layout()  # Ajusta el diseño para evitar solapamiento
plt.show()
```
