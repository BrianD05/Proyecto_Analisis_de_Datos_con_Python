```python
import os
import pandas as pd
import matplotlib.pyplot as plt
# Obtiene el directorio actual
directorio_actual = os.getcwd()

# Nombre del archivo que se desea abrir
nombre_archivo = "Energia-alternativas.xlsx"

# Se Construye la ruta completa del archivo utilizando os.path.join
ruta_completa = os.path.join(directorio_actual, nombre_archivo)

# Verifico si el archivo existe antes de intentar abrirlo
if os.path.exists(ruta_completa):
    # Abre el archivo Excel utilizando pandas
    df = pd.read_excel(ruta_completa)
else:
    print(f"El archivo {nombre_archivo} no existe en el directorio actual.")
```

# Exploracion de Datos 

¿Cuál es la tendencia general de la generación de energía biogás a lo largo de los años?

```python
# Convierto la columna 'indice_tiempo' a tipo datetime
df['indice_tiempo'] = pd.to_datetime(df['indice_tiempo'])

# Filtro para incluir solamente la generación de energía biogás y trabajo con una copia
df_biogas = df[df['actividad_producto_nombre'] == 'Energía biogás'].copy()

# Extrae el año de la columna 'indice_tiempo' del DataFrame filtrado
df_biogas['anio'] = df_biogas['indice_tiempo'].dt.year

# Agrupa los datos por año y calcula la generación total de energía biogás para cada año
generacion_biogas_por_anio = df_biogas.groupby('anio')['valor'].sum()

# Muestra los resultados
print(generacion_biogas_por_anio)
```

```python
# Grafica lineal de la tendencia general del Biogás a los largos de los años
generacion_biogas_por_anio = df_biogas.groupby(df_biogas['indice_tiempo'].dt.year)['valor'].sum()
generacion_biogas_por_anio.plot(kind='line', figsize=(10, 6), marker='o', color='b')
plt.title('Variación de la generación de energía biogás a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Generación de energía biogás (MWh)')
plt.grid(True)
plt.show()
```

```python
# Grafica la tendencia general del Biogás
generacion_biogas_por_anio.plot(kind='bar',figsize=(12,6))
plt.xlabel('Año')
plt.ylabel('Generación de Energía Biogás (MWh)')

plt.title('Tendencia General del Biogás')
plt.show()
```

¿Cuál es la distribución de la generación de energía biogás por provincia?

```python
# Filtro para incluir solamente la generación de energía biogás y trabajar con una copia
df_biogas = df[df['actividad_producto_nombre'] == 'Energía biogás'].copy()

# Calculo la distribución de la energía biogás por alcance del dato
distribucion_energia_biogas = df_biogas['alcance_nombre'].value_counts()

# Mostrar los resultados
print(distribucion_energia_biogas)
```

```python
#DatoAdicional sobre la distribucion de las distintas energia segun su provincia
# Creo una lista de las diferentes actividades/productos de energía
actividades_energia = df['actividad_producto_nombre'].unique()

# Defino el número de filas y columnas para los subplots
num_filas = 2
num_columnas = len(actividades_energia) // num_filas + (len(actividades_energia) % num_filas > 0)

# Creo una figura y ejes para los subplots
fig, axes = plt.subplots(num_filas, num_columnas, figsize=(18, 8))

# Itera sobre las actividades de energía y crea un gráfico de barras para cada una
for i, actividad in enumerate(actividades_energia):
    fila = i // num_columnas
    columna = i % num_columnas
    
    # Filtro el DataFrame para incluir solo la actividad de energía actual
    df_actividad = df[df['actividad_producto_nombre'] == actividad]
    
    # Calcula la distribución de la energía por provincia
    distribucion_energia = df_actividad['alcance_nombre'].value_counts()
    
    # Crea un gráfico de barras para la distribución de la energía por provincia
    distribucion_energia.plot(kind='bar', ax=axes[fila, columna])
    axes[fila, columna].set_title(f'Distribución de la Energía - {actividad}')
    axes[fila, columna].set_xlabel('Provincia')
    axes[fila, columna].set_ylabel('Cantidad')
    axes[fila, columna].grid(axis='y')

# Ajusta el diseño de los subplots
plt.tight_layout()

# Muestra el gráfico
plt.show()
```

```python
#DatoAdicional veamos en forma de tabla

# Calculo la distribución de las energías por actividad y alcance del dato
distribucion_energias = df.groupby(['actividad_producto_nombre', 'alcance_nombre']).size().unstack(fill_value=0)

# Imprimo la distribución de las energías como una tabla
print(distribucion_energias)
```

¿Cuál es la provincia que ha generado más energía biogás en un año específico?

```python
# Convierto la columna 'indice_tiempo' a tipo datetime
df['indice_tiempo'] = pd.to_datetime(df['indice_tiempo'])

# Filtro para incluir solamente la generación de energía biogás y trabajo con una copia
df_biogas = df[df['actividad_producto_nombre'] == 'Energía biogás'].copy()

# Extrae el año de la columna 'indice_tiempo' del DataFrame filtrado
df_biogas['anio'] = df_biogas['indice_tiempo'].dt.year

# Agrupo los datos por año y provincia, y sumo la cantidad de energía biogás generada en cada provincia para cada año
generacion_biogas_por_provincia = df_biogas.groupby(['anio', 'alcance_nombre'])['valor'].sum()

# Encontrar la provincia que ha generado más energía biogás en un año específico (por ejemplo, 2020)
anio_especifico = 2020
if anio_especifico in generacion_biogas_por_provincia.index:
    generacion_biogas_2020 = generacion_biogas_por_provincia.loc[anio_especifico]
    provincia_biogas = generacion_biogas_2020.idxmax()
    cantidad_biogas = generacion_biogas_2020.max()

    print(f"En el año {anio_especifico}, la provincia que ha generado más energía biogás es: {provincia_biogas}, con una cantidad de {cantidad_biogas} MWh.")
    
else:
    
    print(f"No hay datos disponibles para el año {anio_especifico}.")
```

¿Cómo varía la generación de energía biogás en diferentes estaciones del año?

```python
# Convierte la columna 'indice_tiempo' a tipo datetime
df['indice_tiempo'] = pd.to_datetime(df['indice_tiempo'])

# Defino una función para obtener la estación del año a partir de una fecha
def obtener_estacion(fecha):
    mes = fecha.month
    if 3 <= mes <= 6:
        return 'Otoño'
    elif 6 <= mes <= 9:
        return 'Invierno'
    elif 9 <= mes <= 12:
        return 'Primavera'
    else:
        return 'Verano'

# Agrego una columna 'estacion' al DataFrame con la estación del año correspondiente a cada fecha
df['Estacion'] = df['indice_tiempo'].apply(obtener_estacion)

# Filtra el DataFrame para incluir solamente la generación de energía biogás
df_biogas = df[df['actividad_producto_nombre'] == 'Energía biogás']

# Agrupa los datos por estación del año y calcula la generación total de energía biogás para cada estación
generacion_biogas_por_estacion = df_biogas.groupby('Estacion')['valor'].sum()

# Imprime los resultados
print(generacion_biogas_por_estacion)
```

```python
# Grafica la generación total de energía biogás por estación del año
generacion_biogas_por_estacion.plot(kind='bar', color='red')

# Configuración de etiquetas y título
plt.xlabel('Estación del Año')
plt.ylabel('Generación de Energía Biogás (MWh)')
plt.title('Generación de Energía Biogás por Estación del Año')

# Mostrar el gráfico
plt.show()
```

¿Cómo varía la generación de energía solar a lo largo de los años?

```python
#Variacion de energia solar a lo largo de los años
df_solar = df[df['actividad_producto_nombre'] == 'Energía solar'].copy()
df_solar['indice_tiempo'] = pd.to_datetime(df_solar['indice_tiempo'])
generacion_solar_por_anio = df_solar.groupby(df_solar['indice_tiempo'].dt.year)['valor'].sum()
generacion_solar_por_anio.plot(kind='line', figsize=(10, 6), marker='o', color='b')
plt.title('Variación de la generación de energía solar a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Generación de energía solar (MWh)')
plt.grid(True)
plt.show()
```

¿Cuál es la provincia que ha instalado más potencia de energía solar en un año específico?

```python
# Convierto la columna 'indice_tiempo' a tipo datetime
df['indice_tiempo'] = pd.to_datetime(df['indice_tiempo'])

# Filtro para incluir solamente la generación de energía solar y trabajo con una copia
df_solar = df[df['actividad_producto_nombre'] == 'Energía solar'].copy()

# Extrae el año de la columna 'indice_tiempo' del DataFrame filtrado
df_solar['anio'] = df_solar['indice_tiempo'].dt.year

# Agrupo los datos por año y provincia, y sumo la cantidad de energía solar generada en cada provincia para cada año
generacion_solar_por_provincia = df_solar.groupby(['anio', 'alcance_nombre'])['valor'].sum()

# Encontrar la provincia que ha generado más energía solar en un año específico (por ejemplo, 2012)
anio_especifico = 2012
if anio_especifico in generacion_solar_por_provincia.index:
    generacion_solar_2020 = generacion_solar_por_provincia.loc[anio_especifico]
    provincia_mas_solar = generacion_solar_2020.idxmax()
    cantidad_mas_solar = generacion_solar_2020.max()
    print(f"En el año {anio_especifico}, la provincia que ha generado más energía solar es: {provincia_mas_solar}, con una cantidad de {cantidad_mas_solar} MWh.")
else:
    print(f"No hay datos disponibles para el año {anio_especifico}.")
```

¿Cómo ha evolucionado la potencia instalada de energía solar en las diferentes provincias?

```python
# Convierte la columna 'indice_tiempo' a tipo datetime
df['indice_tiempo'] = pd.to_datetime(df['indice_tiempo'])

# Filtra para incluir solamente la generación de energía solar y trabaja con una copia
df_solar = df[df['actividad_producto_nombre'] == 'Energía solar'].copy()

# Agrupa los datos por provincia y año, y calcula la potencia instalada total de energía solar por provincia y año
potencia_instalada_solar_por_provincia_y_anio = df_solar.groupby(['alcance_nombre', df_solar['indice_tiempo'].dt.year])['valor'].sum()

potencia_instalada_solar_por_provincia_y_anio_sin_total = potencia_instalada_solar_por_provincia_y_anio.unstack(level=0)

# Genera la gráfica de barras apiladas
potencia_instalada_solar_por_provincia_y_anio_sin_total.plot(kind='bar', stacked=True, figsize=(12, 8))

# Ajusta las etiquetas y el título
plt.title('Potencia Instalada de Energía Solar por Provincia y Año')
plt.xlabel('Año')
plt.ylabel('Potencia Instalada (MW)')
plt.xticks(rotation=45)  # Rotación de etiquetas en el eje x para una mejor visualización

# Muestra la leyenda fuera del gráfico
plt.legend(title='Provincia', bbox_to_anchor=(1.05, 1), loc='upper left')

# Muestra la gráfica
plt.tight_layout()
plt.show()
```

¿Hay alguna relación entre la potencia instalada de energía solar y la generación de energía biogás?

```python
# Filtro los datos para incluir solamente la potencia instalada de energía solar y la generación de energía biogás
df_solar = df[df['actividad_producto_nombre'] == 'Energía solar']
df_biogas = df[df['actividad_producto_nombre'] == 'Energía biogás']

# Agrupo los datos por año y calcular la suma de la potencia instalada de energía solar y la generación de energía biogás para cada año
potencia_solar_por_anio = df_solar.groupby(df_solar['indice_tiempo'].dt.year)['valor'].sum()
generacion_biogas_por_anio = df_biogas.groupby(df_biogas['indice_tiempo'].dt.year)['valor'].sum()

# Calculo la correlación entre la potencia instalada de energía solar y la generación de energía biogás
correlacion = potencia_solar_por_anio.corr(generacion_biogas_por_anio)

print("Correlación entre potencia instalada de energía solar y generación de energía biogás:", correlacion)
```
